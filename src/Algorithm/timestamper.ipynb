```python
# Opens/creates attendance.csv
# Appends rows: timestamp, name
import csv
from datetime import datetime

CSV_FILE = "attendance.csv"

def log_attendance(name):
    timestamp = datetime.now().isoformat()
    with open(CSV_FILE, 'a', newline='') as f:
        writer = csv.writer(f)
        writer.writerow([timestamp, name])

# Example Usage
log_attendance("Alice")

#https://docs.python.org/3/library/csv.html
#https://docs.python.org/3/library/datetime.html

```

```python
last_logged = {}

def should_log(name):
    now = datetime.now()
    if name not in last_logged or (now - last_logged[name]).seconds > 60:
        last_logged[name] = now
        return True
    return False

def process_frame(frame):
    result = recognize_face(frame)
    if result["name"] and result["confidence"] > 0.8:
        if should_log(result["name"]):
            log_attendance(result["name"])
```
