```python
# Opens/creates attendance.csv
# Appends rows: timestamp,name
import csv
from datetime import datetime
from pathlib import Path

#  Define CSV_FILE 
CSV_FILE = Path(__file__).parent / "attendance.csv"


def initialize_csv():
    """Create CSV with headers if it doesn't exist."""
    if not CSV_FILE.exists():
        with open(CSV_FILE, 'w', newline='') as f:
            writer = csv.writer(f)
            writer.writerow(['timestamp', 'name']) 
            
def log_attendance(name):
    """Log attendance with timestamp."""
    # Create directory if needed
    CSV_FILE.parent.mkdir(parents=True, exist_ok=True)
    
    timestamp = datetime.now().isoformat()
    with open(CSV_FILE, 'a', newline='') as f:
        writer = csv.writer(f)
        writer.writerow([timestamp, name])
    
    print(f"âœ… Logged: {name}")
    print(f"ðŸ“ Location: {CSV_FILE}")

if __name__ == "__main__":
    log_attendance("Alice")





```

```python
last_logged = {}

def should_log(name):
    now = datetime.now()
    if name not in last_logged or (now - last_logged[name]).seconds > 60:
        last_logged[name] = now
        return True
    return False

def process_frame(frame):
    result = recognize_face(frame)
    if result["name"] and result["confidence"] > 0.8:
        if should_log(result["name"]):
            log_attendance(result["name"])
```
